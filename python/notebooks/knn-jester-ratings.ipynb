```python
X = df_init.drop([2],axis=1)
```

```python
y = df_init[2]
```

```python
from sklearn import neighbors
from sklearn.cross_validation import KFold
from sklearn.metrics import mean_squared_error
```

```python
def k_fold_regression(regressor, data, labels, n_folds=5):
    k_fold = KFold(n=data.shape[0], n_folds=n_folds)
    scores = []
    for train_indices, test_indices in k_fold:

        train_text = data[train_indices]
        train_y = labels[train_indices]

        test_text = data[test_indices]
        test_y = labels[test_indices]

        regressor.fit(train_text, train_y)
        predictions = regressor.predict(test_text)
        score = mean_squared_error(test_y,predictions)
        scores.append(score)


    print'MSE:', sum(scores)/len(scores)
```

```python
clf = neighbors.KNeighborsRegressor(1)
k_fold_regression(clf,X.values,y)
```

```python
for k in range(1,4):
    clf = neighbors.KNeighborsRegressor(k,metric='euclidean')
    k_fold_regression(clf,X.values,y)
```
